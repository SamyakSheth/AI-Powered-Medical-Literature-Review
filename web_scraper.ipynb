```python
from bs4 import BeautifulSoup
import requests
import re

# Step 3: Fetch the Web Page
url = 'https://pubmed.ncbi.nlm.nih.gov/?term=Chronic%20illness&sort&&page=1'
response = requests.get(url)

# Step 4: Check Response Status
if response.status_code == 200:
    # Step 5: Parse the HTML
    soup = BeautifulSoup(response.content, 'html.parser')
```

```python
pages = soup.find('label',class_= 'of-total-pages').get_text(strip=True)
pattern = r'\D'
pages = int(re.sub(pattern, '', pages))
pages
```

```python
search_results = soup.find_all('a',class_='docsum-title')
search_results
```

```python
paper_url = []
len(search_results)
for res in search_results:
    href = res.get('href')
    paper_url.append(href)
    print(href)
```

```python
for paper in paper_url: 
    dummy_url = 'https://pubmed.ncbi.nlm.nih.gov'+paper
    print(dummy_url)
```

```python
# Find all <h2> elements with the class 'title'
abstract_headings = soup.find_all('div', class_='abstract-content selected')
abstract_content = abstract_headings[0].find_all('p')
```

```python
formatted_text = '\n\n'.join([p.get_text(strip=True) for p in abstract_content])
print(formatted_text)
```

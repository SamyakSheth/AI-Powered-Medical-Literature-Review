```python
import matplotlib.pyplot as plt

# Sample data for 5 experiments (replace with your actual data)
experiments = ['LLama3.1 Vanilla Pipeline', 'Gemma2 Vanilla Pipeline', 'LLama3 Vanilla Pipeline', 'LLama3 RAG Prompt set1', 'LLama3 RAG Prompt set 2']
TP_values = [0, 7, 5, 18, 23]  # Replace with your actual TP values
FN_values = [30, 23, 25, 12, 7]  # Replace with your actual FN values

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(experiments, TP_values, marker='o', label='True Positives (TP)', color='green')
plt.plot(experiments, FN_values, marker='o', label='False Negatives (FN)', color='red')

# Rotate x-axis labels to avoid overlap
plt.xticks(rotation=45, ha='right')

# Adding titles and labels
plt.title('True Positives and False Negatives Across Experiments')
plt.xlabel('Experiments')
plt.ylabel('Number of TP / FN')
plt.legend()

# Display the plot
plt.grid(True)
plt.show()
```
